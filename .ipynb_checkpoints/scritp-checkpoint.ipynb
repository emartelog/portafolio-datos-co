```python
import pandas as pd
from sqlalchemy import create_engine, text
from dotenv import load_dotenv
import os

load_dotenv("config/.env")

engine = create_engine(
    f"postgresql+psycopg2://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}"
    f"@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)

# Verificar conexión
with engine.connect() as conn:
    resultado = conn.execute(text('SELECT COUNT(*) FROM "GEIH_CARACTERISTICAS_GENERALES"'))
    print(f"✅ Conectado — {resultado.fetchone()[0]:,} registros en características generales")
```

```python
query = """
SELECT 
    f."MES",
    f."DPTO" AS departamento,
    f."FEX_C18" AS factor_expansion,
    f."PET",
    f."P6240" AS actividad,
    CASE 
        WHEN f."P6240"::numeric IN (1.0, 2.0) THEN 'Ocupado'
        WHEN f."P6240"::numeric = 3.0         THEN 'Desocupado'
        WHEN f."P6240"::numeric IN (4.0, 5.0, 6.0) THEN 'Inactivo'
    END AS condicion_actividad
FROM "FUERZA_DE_TRABAJO_GEIH" f
"""

df_ft = pd.read_sql(query, engine)

df_ft['factor_expansion'] = df_ft['factor_expansion'].astype(float)

indicadores = df_ft.groupby(['MES', 'condicion_actividad'])['factor_expansion'].sum().unstack(fill_value=0)
indicadores['PEA'] = indicadores['Ocupado'] + indicadores['Desocupado']
indicadores['TD'] = indicadores['Desocupado'] / indicadores['PEA'] * 100
indicadores['TO'] = indicadores['Ocupado'] / indicadores['PEA'] * 100

print("Indicadores nacionales por mes 2024:")
print(indicadores[['Ocupado','Desocupado','Inactivo','TD','TO']].round(2).to_string())
```

```python
# Departamentos región Caribe
# 08=Atlántico, 13=Bolívar, 20=Cesar, 23=Córdoba, 44=La Guajira, 47=Magdalena, 70=Sucre
caribe = ['08','13','20','23','44','47','70']

nombres = {
    '08':'Atlántico','13':'Bolívar','20':'Cesar',
    '23':'Córdoba','44':'La Guajira','47':'Magdalena','70':'Sucre'
}

query = """
SELECT f."MES", f."DPTO", f."FEX_C18" AS factor_expansion,
    CASE 
        WHEN f."P6240"::numeric IN (1.0,2.0) THEN 'Ocupado'
        WHEN f."P6240"::numeric = 3.0 THEN 'Desocupado'
        WHEN f."P6240"::numeric IN (4.0,5.0,6.0) THEN 'Inactivo'
    END AS condicion
FROM "FUERZA_DE_TRABAJO_GEIH" f
WHERE f."PET"::numeric = 1
AND f."CLASE"::numeric = 1
AND f."DPTO" IN ('08','13','20','23','44','47','70')
"""

df_caribe = pd.read_sql(query, engine)
df_caribe['factor_expansion'] = df_caribe['factor_expansion'].astype(float)
df_caribe['DPTO'] = df_caribe['DPTO'].map(nombres)

ind_caribe = df_caribe.groupby(['DPTO','condicion'])['factor_expansion'].sum().unstack(fill_value=0)
ind_caribe['PEA'] = ind_caribe['Ocupado'] + ind_caribe['Desocupado']
ind_caribe['TD']  = (ind_caribe['Desocupado'] / ind_caribe['PEA'] * 100).round(1)
ind_caribe['TO']  = (ind_caribe['Ocupado'] / (ind_caribe['PEA'] + ind_caribe['Inactivo']) * 100).round(1)
ind_caribe['TGP'] = (ind_caribe['PEA'] / (ind_caribe['PEA'] + ind_caribe['Inactivo']) * 100).round(1)

print("Indicadores región Caribe — año completo 2024:")
print(ind_caribe[['TD','TO','TGP']].sort_values('TD', ascending=False).to_string())
```

```python
import numpy as np

df_tn = pd.read_excel(
    r'C:\Users\efrai\Downloads\anex-GEIH-dic2024.xlsx',
    sheet_name='Total nacional',
    header=None
)

# Encontrar fila de TD
fila_td = df_tn[df_tn[0] == 'Tasa de Desocupación (TD)'].index[0]

# Extraer valores numéricos de esa fila
td_serie = df_tn.iloc[fila_td, 1:].apply(pd.to_numeric, errors='coerce').dropna()

# Tomar los últimos 12 (año 2024 completo)
td_2024 = td_serie.values[-12:]

meses_2024 = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']

df_comparacion = pd.DataFrame({
    'Mes': meses_2024,
    'TD_DANE_oficial': np.round(td_2024.astype(float), 1),
    'TD_nuestro_calculo': [12.6,14.6,15.0,15.6,15.3,14.6,14.6,15.5,16.5,15.4,15.8,12.5]
})

df_comparacion['diferencia'] = (
    df_comparacion['TD_nuestro_calculo'] - df_comparacion['TD_DANE_oficial']
).round(1)

print(df_comparacion.to_string(index=False))
```

```python
# Calcular trimestres móviles con nuestros datos
td_mensual = [12.6,14.6,15.0,15.6,15.3,14.6,14.6,15.5,16.5,15.4,15.8,12.5]
td_oficial =  [12.7,11.7,11.3,10.6,10.3,10.3, 9.9, 9.7, 9.1, 9.1, 8.2, 9.1]

td_trimovil = []
for i in range(len(td_mensual)):
    if i < 2:
        td_trimovil.append(None)
    else:
        trimovil = round((td_mensual[i-2] + td_mensual[i-1] + td_mensual[i]) / 3, 1)
        td_trimovil.append(trimovil)

meses = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']

df_conciliacion = pd.DataFrame({
    'Mes': meses,
    'TD_DANE_oficial': td_oficial,
    'TD_nuestro_mensual': td_mensual,
    'TD_nuestro_trimovil': td_trimovil,
})

df_conciliacion['dif_trimovil'] = (
    pd.to_numeric(df_conciliacion['TD_nuestro_trimovil'], errors='coerce') - 
    df_conciliacion['TD_DANE_oficial']
).round(1)

print(df_conciliacion.to_string(index=False))
```

```python
# ¿Cuántas personas (expandidas) tienes por mes?
query = """
SELECT f."MES",
       SUM(f."FEX_C18"::numeric) AS poblacion_expandida
FROM "FUERZA_DE_TRABAJO_GEIH" f
WHERE f."PET"::numeric = 1
AND f."CLASE"::numeric = 1
GROUP BY f."MES"
ORDER BY f."MES"
"""
df_pet = pd.read_sql(query, engine)
df_pet['poblacion_expandida_mill'] = (df_pet['poblacion_expandida'] / 1000000).round(2)

# Población PET oficial DANE 2024 (en millones, cabecera urbana aprox.)
pet_dane = [25.4, 13.1, 12.6, 12.7, 12.7, 12.6, 12.5, 12.4, 12.3, 12.1, 12.0, 11.9]

df_pet['PET_DANE_mill'] = pet_dane
df_pet['ratio'] = (df_pet['poblacion_expandida_mill'] / df_pet['PET_DANE_mill']).round(2)

print(df_pet.to_string(index=False))
```

```python
# Sin filtrar por CLASE — usar el factor tal como viene
query = """
SELECT f."MES",
       f."FEX_C18" AS factor_expansion,
       CASE 
           WHEN f."P6240"::numeric IN (1.0,2.0) THEN 'Ocupado'
           WHEN f."P6240"::numeric = 3.0 THEN 'Desocupado'
           WHEN f."P6240"::numeric IN (4.0,5.0,6.0) THEN 'Inactivo'
       END AS condicion
FROM "FUERZA_DE_TRABAJO_GEIH" f
WHERE f."PET"::numeric = 1
"""

df_tn2 = pd.read_sql(query, engine)
df_tn2['factor_expansion'] = df_tn2['factor_expansion'].astype(float)

ind_tn2 = df_tn2.groupby(['MES','condicion'])['factor_expansion'].sum().unstack(fill_value=0)
ind_tn2['PEA'] = ind_tn2['Ocupado'] + ind_tn2['Desocupado']
ind_tn2['TD']  = (ind_tn2['Desocupado'] / ind_tn2['PEA'] * 100).round(1)

td_oficial = [12.7,11.7,11.3,10.6,10.3,10.3,9.9,9.7,9.1,9.1,8.2,9.1]
meses = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']

df_final = pd.DataFrame({
    'Mes': meses,
    'TD_DANE_oficial': td_oficial,
    'TD_nuestro': ind_tn2['TD'].values,
    'diferencia': (ind_tn2['TD'].values - td_oficial).round(1)
})

print(df_final.to_string(index=False))
```

```python
query = """
SELECT f."MES",
       f."FEX_C18" AS factor_expansion,
       CASE 
           WHEN f."P6240"::numeric IN (1.0,2.0) THEN 'Ocupado'
           WHEN f."P6240"::numeric = 3.0 THEN 'Desocupado'
           WHEN f."P6240"::numeric IN (4.0,5.0,6.0) THEN 'Inactivo'
       END AS condicion
FROM "FUERZA_DE_TRABAJO_GEIH" f
WHERE f."PET"::numeric = 1
AND f."PER"::numeric = 2024
"""

df_tn4 = pd.read_sql(query, engine)
df_tn4['factor_expansion'] = df_tn4['factor_expansion'].astype(float)

# Calcular TD por mes individualmente
resultados = []
for mes in sorted(df_tn4['MES'].unique()):
    df_mes = df_tn4[df_tn4['MES'] == mes]
    g = df_mes.groupby('condicion')['factor_expansion'].sum()
    ocupados    = g.get('Ocupado', 0)
    desocupados = g.get('Desocupado', 0)
    inactivos   = g.get('Inactivo', 0)
    pea = ocupados + desocupados
    td  = round(desocupados / pea * 100, 1)
    to  = round(ocupados / (pea + inactivos) * 100, 1)
    tgp = round(pea / (pea + inactivos) * 100, 1)
    resultados.append({'MES': mes, 'TD': td, 'TO': to, 'TGP': tgp})

df_ind = pd.DataFrame(resultados)

td_oficial = [12.7,11.7,11.3,10.6,10.3,10.3,9.9,9.7,9.1,9.1,8.2,9.1]
meses = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']

df_ind['Mes'] = meses
df_ind['TD_DANE'] = td_oficial
df_ind['diferencia'] = (df_ind['TD'] - df_ind['TD_DANE']).round(1)

print(df_ind[['Mes','TD_DANE','TD','TO','TGP','diferencia']].to_string(index=False))
```

```python
query = """
SELECT f."MES", f."P6240", COUNT(*) AS personas
FROM "FUERZA_DE_TRABAJO_GEIH" f
WHERE f."PET"::numeric = 1
AND f."PER"::numeric = 2024
GROUP BY f."MES", f."P6240"
ORDER BY f."MES", f."P6240"
"""
df_dist = pd.read_sql(query, engine)

# Pivotear para ver la distribución por mes
df_pivot = df_dist.pivot(index='MES', columns='P6240', values='personas').fillna(0)
df_pivot.columns = [f'cod_{c}' for c in df_pivot.columns]
df_pivot['total'] = df_pivot.sum(axis=1)
df_pivot['pct_cod3'] = (df_pivot['cod_3.0'] / df_pivot['total'] * 100).round(1)

print("Porcentaje de desocupados (cod 3) por mes:")
print(df_pivot[['cod_3.0','total','pct_cod3']].to_string())
```

```python
query = """
SELECT f."MES", f."FEX_C18" AS factor_expansion,
    CASE 
        WHEN f."P6240"::numeric IN (1.0,2.0) THEN 'Ocupado'
        WHEN f."P6240"::numeric = 3.0 THEN 'Desocupado'
        WHEN f."P6240"::numeric IN (4.0,5.0,6.0) THEN 'Inactivo'
    END AS condicion
FROM "FUERZA_DE_TRABAJO_GEIH" f
WHERE f."PET"::numeric = 1
AND f."CLASE"::numeric = 1
AND f."DPTO" = '13'
"""

df_bolivar = pd.read_sql(query, engine)
df_bolivar['factor_expansion'] = df_bolivar['factor_expansion'].astype(float)

ind_bolivar = df_bolivar.groupby(['MES','condicion'])['factor_expansion'].sum().unstack(fill_value=0)
ind_bolivar['PEA'] = ind_bolivar['Ocupado'] + ind_bolivar['Desocupado']
ind_bolivar['TD']  = (ind_bolivar['Desocupado'] / ind_bolivar['PEA'] * 100).round(1)
ind_bolivar['TO']  = (ind_bolivar['Ocupado'] / (ind_bolivar['PEA'] + ind_bolivar['Inactivo']) * 100).round(1)

print("Bolívar — mercado laboral mes a mes 2024:")
print(ind_bolivar[['TD','TO']].to_string())
```
