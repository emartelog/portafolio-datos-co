# Análisis de mercado laboral — GEIH
**Fuente:** DANE — Gran Encuesta Integrada de Hogares  
**Procesamiento:** Python + PostgreSQL  

Este notebook calcula indicadores de mercado laboral para Colombia y la región Caribe.
Se adapta automáticamente al año más reciente disponible en la base de datos.

```python
# ── Librerías y conexión ──────────────────────────────────────
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from sqlalchemy import create_engine, text
from dotenv import load_dotenv
from pathlib import Path
import os

load_dotenv('config/.env')
engine = create_engine(
    f"postgresql+psycopg2://{os.getenv('DB_USER')}:{os.getenv('DB_PASSWORD')}"
    f"@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)

# Verificar conexión
with engine.connect() as conn:
    n = conn.execute(text('SELECT COUNT(*) FROM "GEIH_CARACTERISTICAS_GENERALES"')).fetchone()[0]
    print(f"Conexión exitosa — {n:,} registros en características generales")
```

```python
# ── Parámetros globales ───────────────────────────────────────
# Cambia ANNO para analizar un año específico.
# None = detecta automáticamente el más reciente con 12 meses completos.
ANNO = None

# Detectar años disponibles en la base
df_annos = pd.read_sql("""
    SELECT DISTINCT "PER"::numeric AS anno,
           COUNT(DISTINCT "MES") AS meses
    FROM "FUERZA_DE_TRABAJO_GEIH"
    WHERE "PET"::numeric = 1
    GROUP BY "PER"::numeric
    ORDER BY anno
""", engine)
df_annos['anno'] = df_annos['anno'].astype(int)

print("Años disponibles:")
print(df_annos.to_string(index=False))

if ANNO is None:
    anno_analisis = int(df_annos[df_annos['meses'] == 12]['anno'].max())
    print(f"\nAño seleccionado automáticamente: {anno_analisis}")
else:
    anno_analisis = ANNO
    print(f"\nAño seleccionado: {anno_analisis}")

# Diccionarios compartidos
MESES_NOMBRES = {
    '01':'Ene','02':'Feb','03':'Mar','04':'Abr',
    '05':'May','06':'Jun','07':'Jul','08':'Ago',
    '09':'Sep','10':'Oct','11':'Nov','12':'Dic'
}

DPTOS_CARIBE = ['08','13','20','23','44','47','70']
NOMBRES_CARIBE = {
    '08':'Atlántico','13':'Bolívar','20':'Cesar',
    '23':'Córdoba','44':'La Guajira','47':'Magdalena','70':'Sucre'
}
```

```python
# ── Función base de indicadores ───────────────────────────────
# Reutilizable para cualquier subconjunto de datos.
# Calcula TD, TO y TGP mes a mes o por agrupación.

def calcular_indicadores(df, agrupar_por):
    """
    Calcula TD, TO y TGP agrupando por las columnas indicadas.
    df debe tener columnas: condicion, factor_expansion, y las de agrupar_por.
    """
    g = df.groupby(agrupar_por + ['condicion'])['factor_expansion'].sum().unstack(fill_value=0)
    g['PEA'] = g.get('Ocupado', 0) + g.get('Desocupado', 0)
    g['TD']  = (g.get('Desocupado', 0) / g['PEA'] * 100).round(1)
    g['TO']  = (g.get('Ocupado', 0) / (g['PEA'] + g.get('Inactivo', 0)) * 100).round(1)
    g['TGP'] = (g['PEA'] / (g['PEA'] + g.get('Inactivo', 0)) * 100).round(1)
    return g[['TD','TO','TGP']]

print("Función calcular_indicadores() lista.")
```

```python
# ── Indicadores nacionales mes a mes ─────────────────────────
# Filtra por PET=1 y el año de análisis.
# No filtra por CLASE para cubrir el total nacional.

df_ft = pd.read_sql(f"""
    SELECT f."MES",
           f."FEX_C18" AS factor_expansion,
           CASE
               WHEN f."P6240"::numeric IN (1.0,2.0) THEN 'Ocupado'
               WHEN f."P6240"::numeric = 3.0         THEN 'Desocupado'
               WHEN f."P6240"::numeric IN (4.0,5.0,6.0) THEN 'Inactivo'
           END AS condicion
    FROM "FUERZA_DE_TRABAJO_GEIH" f
    WHERE f."PET"::numeric = 1
    AND   f."PER"::numeric = {anno_analisis}
""", engine)
df_ft['factor_expansion'] = df_ft['factor_expansion'].astype(float)

ind_nac = calcular_indicadores(df_ft, ['MES'])
ind_nac.index = ind_nac.index.map(MESES_NOMBRES)

print(f"Indicadores nacionales — {anno_analisis}")
print(ind_nac.to_string())
```

```python
# ── Indicadores región Caribe — año completo ─────────────────
# Un solo valor por departamento promediando todo el año.

dptos_str = "','".join(DPTOS_CARIBE)

df_caribe = pd.read_sql(f"""
    SELECT f."MES", f."DPTO",
           f."FEX_C18" AS factor_expansion,
           CASE
               WHEN f."P6240"::numeric IN (1.0,2.0) THEN 'Ocupado'
               WHEN f."P6240"::numeric = 3.0         THEN 'Desocupado'
               WHEN f."P6240"::numeric IN (4.0,5.0,6.0) THEN 'Inactivo'
           END AS condicion
    FROM "FUERZA_DE_TRABAJO_GEIH" f
    WHERE f."PET"::numeric = 1
    AND   f."PER"::numeric = {anno_analisis}
    AND   f."DPTO" IN ('{dptos_str}')
""", engine)
df_caribe['factor_expansion'] = df_caribe['factor_expansion'].astype(float)
df_caribe['DPTO'] = df_caribe['DPTO'].map(NOMBRES_CARIBE)

ind_caribe = calcular_indicadores(df_caribe, ['DPTO'])
ind_caribe = ind_caribe.sort_values('TD', ascending=False)

print(f"Indicadores región Caribe — año completo {anno_analisis}")
print(ind_caribe.to_string())
```

```python
# ── Bolívar — tendencia mensual ───────────────────────────────
# Zoom en Bolívar (DPTO=13) para ver el comportamiento mes a mes.

df_bolivar = df_caribe[df_caribe['DPTO'] == 'Bolívar'].copy()
ind_bolivar = calcular_indicadores(df_bolivar, ['MES'])
ind_bolivar.index = ind_bolivar.index.map(MESES_NOMBRES)

print(f"Bolívar — mercado laboral mes a mes {anno_analisis}")
print(ind_bolivar.to_string())
```

```python
# ── Gráfica: Bolívar vs promedio nacional ─────────────────────
# Muestra la brecha entre Bolívar y la TD nacional.
# La línea de referencia se actualiza automáticamente con el año.

# TD nacional de diciembre (último mes del año analizado)
td_dic_nacional = ind_nac.loc['Dic', 'TD'] if 'Dic' in ind_nac.index else None

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=ind_bolivar.index,
    y=ind_bolivar['TD'],
    mode='lines+markers',
    name='Tasa de desocupación — Bolívar',
    line=dict(color='#E8593C', width=2.5),
    marker=dict(size=8)
))

fig.add_trace(go.Scatter(
    x=ind_bolivar.index,
    y=ind_bolivar['TO'],
    mode='lines+markers',
    name='Tasa de ocupación — Bolívar',
    line=dict(color='#1D9E75', width=2.5),
    marker=dict(size=8)
))

# Línea de referencia nacional
if td_dic_nacional:
    fig.add_hline(
        y=td_dic_nacional,
        line_dash='dash',
        line_color='gray',
        annotation_text=f"TD nacional dic-{anno_analisis} ({td_dic_nacional}%)",
        annotation_position='bottom right'
    )

fig.update_layout(
    title=f'Mercado laboral — Bolívar {anno_analisis}',
    xaxis_title='Mes',
    yaxis_title='Tasa (%)',
    template='plotly_white',
    legend=dict(orientation='h', yanchor='bottom', y=1.02),
    height=450
)

fig.show()
```

```python
# ── Gráfica: comparativo región Caribe ───────────────────────
# Barras horizontales con la TD de cada departamento.
# Resalta Bolívar como departamento de interés.

colores = [
    '#E8593C' if d == 'Bolívar' else '#B4B2A9'
    for d in ind_caribe.index
]

fig2 = go.Figure(go.Bar(
    x=ind_caribe['TD'],
    y=ind_caribe.index,
    orientation='h',
    marker_color=colores,
    text=ind_caribe['TD'].astype(str) + '%',
    textposition='outside'
))

fig2.update_layout(
    title=f'Tasa de desocupación — Región Caribe {anno_analisis}',
    xaxis_title='TD (%)',
    yaxis_title='',
    template='plotly_white',
    height=400,
    yaxis={'categoryorder': 'total ascending'}
)

fig2.show()
```

```python
# ── Guardar resultados ────────────────────────────────────────
os.makedirs('datos/processed', exist_ok=True)

ind_nac.to_csv(f'datos/processed/indicadores_nacionales_{anno_analisis}.csv')
ind_caribe.to_csv(f'datos/processed/indicadores_caribe_{anno_analisis}.csv')
ind_bolivar.to_csv(f'datos/processed/indicadores_bolivar_{anno_analisis}.csv')

print(f"Resultados guardados en datos/processed/ para el año {anno_analisis}")
```
